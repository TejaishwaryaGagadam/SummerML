```
import numpy as np
import matplotlib.pyplot as plt
```

```
def print_and_plot(x,y):
  plt.figure()
  plt.plot(x,y)
  plt.grid()
  print('Mean = {}'.format(np.mean(y)))
  print('Variance = {}'.format(np.var(y)))
```

```
x1 = np.array([0,2,5,4])
y1 = np.array([1,3,2,1])
plt.figure()
plt.plot(x1,y1,'o')
plt.grid()
print('Mean = {}'.format(np.mean(y1)))
print('Variance = {}'.format(np.var(y1)))
```

```
x = np.linspace(-1,1,100)
print(x)
y2 = 2*x + 0.5
print_and_plot(x,y2)
```

```
y3 = np.sin(2*np.pi*x)
print_and_plot(x,y3)
```

```
y4 = x**3 + 2
print_and_plot(x,y4)
```

```
y5 = np.exp(-2*x)
print_and_plot(x,y5)
```

```
sigma = 0.5
mean = 0
y6 = 1/np.sqrt(2*np.pi*sigma**2) * np.exp(-2*(x-mean)**2/(2*sigma**2))
print_and_plot(x,y6)
```

```
## All-in-one plot


plt.figure()  #figsize = (8,8)
plt.plot(x,y2)  # c = 'green'  change color
plt.plot(x,y3)
plt.plot(x,y4)
plt.plot(x,y5)
plt.plot(x,y6)

plt.plot(x1,y1,'o')
## After add scatter points, need to change x to (-1,5) to get better plot
plt.ylim([-2,4])
plt.xlim([-1,5])
plt.grid()
plt.legend(['Linear','Sine','Poly','Expo','Gauss','Scatter'])
```
